## Preparing the Dataset

```python
dataset = pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
dataset.head()
```

```python
dataset["Price"] = boston.target
```

```python
dataset.head()
```

```python
# Summarizing the stats of the data
dataset.describe()
```

```python
# Check the missing values(Super Imp Step)
# dataset.isnull()
dataset.isnull().sum()
```

```python
### Exploratory Data Analysis
## Correlation (Super Important for Reguarisation Problems)
# It tells us how the features are correleated

# Making our  correlation matrix a liitle prettier
corr_matrix = dataset.corr() 
fig, ax = plt.subplots(figsize=(15, 10))

import seaborn as sns

ax = sns.heatmap(corr_matrix, 
                     annot= True,
                     linewidths= 0.5,
                     fmt= ".2f",
                     cmap= "YlGnBu")
```

```python
## Independent and Dependent Features

x = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
x.head()
```

```python
y
```

```python
## Train, Test Split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

## Model Training

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
```

```python
regression = LinearRegression()
sgd_regression = SGDRegressor()
```

```python
sgd_regression.fit(x_train, y_train)
```

```python
regression.fit(x_train, y_train)
```

```python
# print the coefficicents and intercept
print(regression.coef_)

# So what does it tell?
# It tells with the inc in feature of how many unit ,how much unit got
# inc/dec in output For ex Suppose with an inc of 1000 unit in crime 
 # price decrease by 946$
```

```python
print(regression.intercept_)
```

```python
## On which params model got trained
regression.get_params()
```

```python
### Prediction with Test data
reg_pred = regression.predict(x_test)
sgd_reg_pred = sgd_regression.predict(x_test)
```

```python
sgd_reg_pred
```

```python
reg_pred
```

## Checking Performance with the help of Plots

```python
# Plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
## Residual plot(residual = error with respect to x and y)
residuals = y_test - reg_pred 
```

```python
residuals
```

```python
# Plot the residuals
sns.displot(residuals, kind='kde')
```

```python
## Scatter plot with respect to prediction and residuals
## UNIFORM distribution
plt.scatter(reg_pred, residuals)
```

### Performance metrices

## R square and Adjusted R Square

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
sgd_score = r2_score(y_test, sgd_reg_pred)
print(score,sgd_score)
```

```python
# Adjusted R square
1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```
